```python
import pandas as pd
```

```python
df = pd.read_csv("Drivers_Input.csv")
```

```python
vibe_map = {1: 'Positive', 2: 'Mixed', 3: 'Negative'}
df['Sentiment'] = df['Vibe'].map(vibe_map)
```

```python
pivot = pd.pivot_table(
    df,
    index='Keyword Notes',
    columns='Vibe',
    values='User Id',
    aggfunc='count',
    fill_value=0
)
```

```python

pivot['Grand Total'] = pivot.sum(axis=1)

```

```python

sentiment_counts = df.groupby(['Keyword Notes', 'Sentiment']).size().unstack(fill_value=0)
sentiment_perc = sentiment_counts.div(sentiment_counts.sum(axis=1), axis=0) * 100
sentiment_perc = sentiment_perc.round().astype(int).astype(str) + '%'
```

```python
final = pivot.merge(sentiment_perc, left_index=True, right_index=True, how='outer').fillna('0%')
```

```python
final = final.sort_values(by='Grand Total', ascending=False)
```

```python
final.to_csv("Drivers_DesiredOutput.csv", index_label="Overall: Drivers")
```
