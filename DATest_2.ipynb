```python
import pandas as pd

```

```python
df = pd.read_csv('Age_Input.csv')
```

```python
df.info()
```

```python
df.describe()
```

```python
pivot = df.pivot_table(index='Age', columns='Keyword', values='User Id', aggfunc='count', fill_value=0)
```

```python
pivot['Grand Total'] = pivot.sum(axis=1)
```

```python
total_row = pivot.sum(numeric_only=True)
total_row.name = 'Grand Total'
```

```python
pivot = pd.concat([pivot, pd.DataFrame([total_row])])
```

```python
percent = pivot.div(pivot.loc['Grand Total'], axis=1)
percent = percent.applymap(lambda x: f"{x:.0%}" if isinstance(x, (int, float)) else x)
```

```python
combined = pd.concat([pivot, percent], axis=1)
combined.columns = [*pivot.columns, *[f"{col} %" for col in pivot.columns]]
```

```python
combined.to_csv('Age_DesiredOutput.csv', index=True)
```
